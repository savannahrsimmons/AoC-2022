```python
input = open('day_11.txt').readlines()

print()
monkey_items = []
# 0 operation type, 1 operation number, 2 test, 3 true, 4 false
monkey_process = []
monkey_inspections = []

i = 0

while i < len(input):
    items = []
    monkey = int(input[i][7])
    monkey_items.append([])
    monkey_process.append([0,0,0,0,0])
    monkey_inspections.append(0)
    
    items = input[i+1].split()
    j = 2
    while j < len(items):
        monkey_items[monkey].append(int(items[j].strip(',')))
        j += 1
        
    operation = input[i+2].split()
    monkey_process[monkey][0] = operation[4]
    if(operation[5] == 'old'):
        monkey_process[monkey][1] = operation[5]
    else:
        monkey_process[monkey][1] = int(operation[5])
    
    monkey_process[monkey][2] = int(input[i+3].split()[3])
    monkey_process[monkey][3] = int(input[i+4].split()[5])
    monkey_process[monkey][4] = int(input[i+5].split()[5])
    
    i += 7

print(monkey_inspections)
    
for round in range (0,20):
    # each monkey
    for monkey in range(0,len(monkey_items)):
        # each item
        for item in monkey_items[monkey]:
            
            # inspection count
            monkey_inspections[monkey] += 1
            
            # operation
            if monkey_process[monkey][1] == 'old':
                number = item
            else:
                number = monkey_process[monkey][1]
            if monkey_process[monkey][0] == '+':
                item += number
            else:
                item *= number
            item = int(item/3)
            
            # test
            if item % monkey_process[monkey][2] == 0:
                monkey_items[monkey_process[monkey][3]].append(item)
            else:
                monkey_items[monkey_process[monkey][4]].append(item)
        #clear out items   
        monkey_items[monkey] = []
            
print(monkey_inspections)
#print(monkey_items)
#print(monkey_process)

monkey_inspections.sort()
print(monkey_inspections[-1] * monkey_inspections[-2])

```
